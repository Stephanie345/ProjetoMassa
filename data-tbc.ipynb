### Três regiões com maiores notificações de turbelose no Brasil (Sudeste, Nordeste e Sul)
As três regiões que concentram maior número de notificações de turbeculose no Brasil são: Região Sudeste (estados de SP e RJ), Região Nordeste (estados PE e BA) e região Sul (estado RS)

### Vamos analisar graficamente a taxa de notificações de turbeculose para os 5 primeiros estados no ranking.

```python
# Plotando gráfico de 2007 á 2017, usando ascending=False (para ordem decrescente) e x=estados de notificação
g= tbc.sort_values(by= ["ano 2007","ano 2008","ano 2009","ano 2010","ano 2011","ano 2012","ano 2013","ano 2014","ano 2015","ano 2016","ano 2017"],ascending=False).head(5).plot(kind="barh", grid=(True), xticks=(5000,7500,10000,12500,15000,17500,20000,22500), x=("Estado de notifição"),figsize=(20,10), fontsize=(16))

g.set_title("Notificações de turbeculose por estados no Brasil", fontsize=(20));
g.set_xlabel("Número de notificações de turbeculose", fontsize=(18));
g.set_ylabel("Estados notificados", fontsize=(18));
```

### Análise dos útlimos 10 anos
Analisando os estados notificados nos últimos 10 anos, pode-se fazer as seguintes observações:
#### Bahia (BA) Maior queda: 
Estado que apresentou maior queda de notificações ao longo dos últimos 10 anos, porém em 2017 tornou a apresentar um número maior de notificações.
#### São Paulo (SP) Maior aumento:
Estado que apresentou maior aumento de notificações ao longo dos últimos 10 anos. Em meados de 2012 houve pequena redução, mas a partir do ano seguinte tornou a apresentar crescimento das notificações.



### Faixa etária de notificações de turbeculos no estado de São Paulo
Serão realizadas as análises da idade em que há maior número de notificações de turbeculose em São Paulo

```python
#carregando planilha de faixa etária para o estado de São Paulo
tbc_idade=pd.read_excel('C:\\Users\\julia\Documents\\00-PYTHON\\Turbeculose-analise-dados\\datasusuf_idade.xlsx', skiprows=5)
tbc_idade.head()
```

```python
#Análise da média de notificações de todas as idades ao longo de 10 anos, ocorrencia maxima e mínima.
tbc_idade.describe()
```

```python
#Visualizando graficamente as notificações por faixa etária
f= tbc_idade.sort_values(by= ["ano 2007","ano 2008","ano 2009","ano 2010","ano 2011","ano 2012","ano 2013","ano 2014","ano 2015","ano 2016","ano 2017"],ascending=False).head().plot(kind="barh", grid=(True), xticks=(600,1200,1800, 2400, 3000, 3600, 4200, 4800, 5400, 6000, 6600, 7200, 7800, 8400, 9000, 9600, 10200, 10800, 11400, 12000), x=("Faixa etária"),figsize=(20,10), fontsize=(16))
f.set_title("Notificações de turbeculose por faixa etária no estado de São Paulo", fontsize=(20));
f.set_xlabel("Número de notificações de turbeculose", fontsize=(18));
f.set_ylabel("Faixa etária", fontsize=(18));
```
