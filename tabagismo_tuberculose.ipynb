### Estados notificados de tuberculos devido ao tabagismo de 2007 a 2017

```python
# Plotando gráfico de 2007 á 2017, usando ascending=False (para ordem decrescente) e x=estados de notificação
a= tbc.sort_values(by= ["ano 2007","ano 2008","ano 2009","ano 2010","ano 2011","ano 2012","ano 2013","ano 2014","ano 2015","ano 2016","ano 2017"],ascending=False).head().plot(kind="barh", grid=(True), xticks=(0,550,1100,1650,2200,2750,3300,3850,4400,4950,5500), x=("Estado de notifição"),figsize=(20,10), fontsize=(16))
a.set_title("Notificações de turbeculose devido ao tabagismo por estados no Brasil ", fontsize=(20));
a.set_xlabel("Número de notificações de turbeculose devido ao tabagismo", fontsize=(18));
a.set_ylabel("Estados notificados", fontsize=(18));
```

### Análise dos útlimos 10 anos
Analisando os estados notificados nos últimos 10 anos, pode-se fazer as seguintes observações:
#### Pará Maior queda: 
Estado que apresentou menor número de casos de tubercolose por tabagismo notificados ao longo dos últimos 10 anos, porém em meados de 2015 tornou a crescer o número da doença.
#### São Paulo (SP) Maior aumento:
Estado que apresentou maior aumento de notificações ao longo dos últimos 10 anos. Em meados de 2010 houve pequena redução, mas a partir do ano seguinte tornou a apresentar crescimento das notificações.
